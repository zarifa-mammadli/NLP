```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd

df = pd.read_table('modified_aze_link_sentences.tsv')
lst = df.iloc[:, 0].tolist()[:5]


list_json = []

for url in lst:
    vstr = requests.get(url).content
    soup = BeautifulSoup(vstr, features="html.parser")

    rows = soup.findAll('div', {"class":"sentence-and-translations md-whiteframe-1dp"})

    for row in rows:
        data = '[' + row['ng-init'][11:-1] + ']'
        new_data = str(data).replace("'und'", '"und"')
        new_data = json.loads(new_data)
        list_json.append(new_data)
                  
list_json
```

```python
sentences = []
for i in range(len(list_json)):
    x = dict()
    x['id'] = list_json[i][0]['id']
    x[list_json[i][0]['lang']] = list_json[i][0]['text']
    x['correctnesss'] = list_json[i][0]['correctness']
    for j in range(0,len(list_json[0][0])):
        x[list_json[i][0]['translations'][0][0]['lang_tag']] = list_json[i][0]['translations'][0][0]['text']
        x[list_json[i][0]['translations'][0][0]['lang_name']] = list_json[i][0]['translations'][0][0]['lang']
        list_json[i][0]['translations'][0][0]['correctness']
        lang_tag = list_json[1][0]['translations'][1][j]['lang_tag']
        text = list_json[1][0]['translations'][1][j]['text']
        x[lang_tag] =text
        x[list_json[1][0]['translations'][1][j]['lang_name']] = list_json[1][0]['translations'][1][j]['lang']
        list_json[1][0]['translations'][1][j]['correctness']
        
        
    sentences.append(x)   
sentences
```

```python
sentences_list = pd.DataFrame(sentences)
sentences_list
```

```python
sentences_list.to_csv('sentences.tsv', sep='\t', index=False)
```
